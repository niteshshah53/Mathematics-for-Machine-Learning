```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Standardize the data (mean=0 and variance=1)
X_standardized = StandardScaler().fit_transform(X)
```

```python
# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 principal components
X_pca = pca.fit_transform(X_standardized)
```

```python
# Plot the original and reduced data
plt.figure(figsize=(12, 6))
```

```python
# Plot original data
plt.subplot(1, 2, 1)
for i, c in zip(range(3), ['red', 'green', 'blue']):
    plt.scatter(X[y == i, 0], X[y == i, 1], c=c, label=f'Class {i}')
plt.title('Original Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
```

```python
# Plot data after PCA
plt.subplot(1, 2, 2)
for i, c in zip(range(3), ['red', 'green', 'blue']):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], c=c, label=f'Class {i}')
plt.title('Data after PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()

plt.tight_layout()
plt.show()
```

# example 2 for PCA

```python
# Step 1: Get your toys ready (import libraries and load data)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Imagine you have a bunch of toys with different features (like colors and sizes)
toys = np.array([[3, 5], [1, 2], [4, 6], [2, 3], [5, 7]])

# Step 2: Look at all the toys and their features
print("All the toys and their features:")
print(toys)

# Step 3: Find the main ideas (Apply PCA to find components)
pca = PCA(n_components=1)  # We want to find 1 main idea/component
main_idea = pca.fit_transform(toys)

# Step 4: Keep the important parts (Transform the data using the main idea)
print("\nKeeping the important parts:")
print(main_idea)

# Step 5: Tell your friends about the important parts
# (Plotting the toys before and after using PCA)
plt.scatter(toys[:, 0], toys[:, 1], color='blue', label='Original Toys')
plt.scatter(main_idea, [0] * len(main_idea), color='red', marker='*', s=200, label='Main Idea')
plt.xlabel('Color')
plt.ylabel('Size')
plt.title('Toys and Their Main Idea')
plt.legend()
plt.show()
```
